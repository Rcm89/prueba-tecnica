```python
# URL y encabezados para la API
url = "https://moviesdatabase.p.rapidapi.com/titles"
headers = {
    "x-rapidapi-key": "f0aa977acemsh42d7af0bca32ac5p147ffejsn74e30ab678c4",
	"x-rapidapi-host": "moviesdatabase.p.rapidapi.com"
}

# Parámetros base
params = {
    "startYear": 1990,
    "endYear": 2023,
    "titleType": "movie,short",
    "genre": "Drama,Comedy,Action,Fantasy,Horror,Mystery,Romance,Thriller",
    "limit": 10,  # Máximo por página según la API
    "sort": "year.decr",
    "info": "mini_info"
}

# Inicializar variables para iterar
all_results = []
page = 1

while True:
    params["page"] = page
    response = requests.get(url, headers=headers, params=params)
    
    if response.status_code != 200:
        print(f"Error: {response.status_code}, {response.text}")
        break

    data = response.json()
    results = data.get('results', [])
    
    if not results:
        # Salir del bucle si no hay más resultados
        print("No hay más resultados.")
        break
    
    # Extraer datos relevantes
    all_results.extend([
        (
            item.get('id', 'N/A'),  # ID de la película
            item.get('titleText', {}).get('text', 'N/A'),  # Título
            item.get('releaseYear', {}).get('year', 'N/A'),  # Año de estreno
            item.get('releaseDate', {}).get('month', 'N/A'),  # Mes de estreno
            item.get('titleType', {}).get('text', 'N/A')  # Tipo
        )
        for item in results
    ])
    
    # Imprimir progreso
    print(f"Página {page} procesada, {len(results)} resultados encontrados.")
    
    # Verificar si hay más páginas
    if not data.get("next"):
        print("No hay más páginas.")
        break
    
    # Incrementar el número de página
    page += 1

# Mostrar todos los resultados
print(f"Se extrajeron {len(all_results)} resultados.")
```
